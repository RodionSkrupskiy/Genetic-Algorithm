```python
import numpy as np
import random
import time
start_time = time.time()

# Booth

mu = 1000000
lmbda  = 100

eps = 1e-8
normalizer = 1e-3
dim = 2

fitness = lambda x: (x[0] +2*x[1] - 7)**2 + (2*x[0] + x[1] - 5)**2

def mutate(individual):
    ind_1 = individual + np.random.randn(2) * 0.001
    ind_2 = individual + np.random.randn(2) * 0.001
    ind_3 = individual + np.random.randn(2) * 0.001
    ind_4 = individual + np.random.randn(2) * 0.001
    ind = np.concatenate([ind_1, ind_2, ind_3, ind_4])
    return ind

def generate_population(amount, dimension):
    return (10 * np.random.rand(amount, dimension) - 5)

def choose_best(ancestors, descendants, amount):
    descendants_fit = np.array(list(map(fitness, descendants)))
    return descendants[descendants_fit.argsort()[:amount]]

def condition(population_prev, population_cur):
    min_value_prev = min(list(map(fitness, population_prev)))
    min_value_cur = min(list(map(fitness, population_cur)))
    return abs(min_value_prev - min_value_cur) < eps


def evolution():
    minimum = [1,3]
    counter = 1
    population = generate_population(mu, dim)
    mutated_population = mutate(choose_best(population, population, lmbda))
    
    minimum[0], minimum[1] = mutated_population[0], fitness(mutated_population[0])

    while not condition(population, mutated_population):
        population = mutated_population
        mutated_population = mutate(choose_best(mutated_population, mutated_population, lmbda))
        minimum[0], minimum[1] = mutated_population[0], fitness(mutated_population[0])
        print('(x1, x2) = {}; F(x1, x2) = {}'.format(minimum[0], minimum[1]))
    print("--- %s seconds ---" % (time.time() - start_time))
    return 'arg = {}; fitness = {}'.format(minimum[0], minimum[1])

evolution()
```

```python
#Genetic algotithm for minimisation of test function
#method of deformed stars
#Booth
import math as m
import numpy as np
import random as rn
import matplotlib.pyplot as plt
import time
start_time = time.time()


dim = 2 
number = 10000

#все эти значения я выбрал рандомно, наверняка есть более оптимальные углы и коэфы
a = 2          #дистанция переноса
alpha = 1 #угол поворота при переносе
beta = 0.5  #угол поворота в шаге 5
k = 2        #коэффициент сжатия
eps = 0.0000001#точность

fitness = lambda x: (x[0] +2*x[1] - 7)**2 + (2*x[0] + x[1] - 5)**2

#Step 1 бесполезный. т.к. точность зависит от eps, а не от итерации
t = 0

#Step 2, 3
def Generate_population_t(number, dim):
    population = 10*np.random.rand(number, dim) - 5
    return population

#Step 4 
def Generate_population_z(population):
    population_z = np.array([])#нужно из начальной популяции создать подобную с таким же к-вом элементов
    while(population_z.size < population.size):#поэтому цикл работает, пока новая популяция не станет по размеру как начальная
        el = rn.choice(population)
        new_el = np.array([el[0] + a*m.cos(alpha), el[1] + a*m.sin(alpha)])#метода
        if abs(new_el[0]) <= 5 and abs(new_el[1]) <= 5:#элементы не должны вылезать за пределы прямоугольника
            population_z = np.concatenate([population_z, new_el])#поэлементно заполняем новую популяцию
    population_z = np.reshape(population_z, (int(population_z.size/2),2))#делим на кусочки по два
    return population_z

#Step 5
def Generate_population_s(population):
    population_s = np.array([])
    while(population_s.size < population.size):
        el_1 = rn.choice(population)
        el_2 = rn.choice(population)
        if fitness(el_1) < fitness(el_2):#если значение ф-ии в точке 1 меньше, чем в т. 2, то мы перезаписываем эту точку
            new_el_i = el_1#в новую переменную, а вторую точку трансформируем формулой из методы
            el_j = el_2#а именно поворачиваем на угол beta
        else:
            new_el_i = el_2
            el_j = el_1
        new_el_j_1 = el_j[0] + (el_j[0] - new_el_i[0])*m.cos(beta) - (el_j[1] - new_el_i[1])*m.sin(beta)
        new_el_j_2 = el_j[1] + (el_j[0] - new_el_i[0])*m.sin(beta) + (el_j[1] - new_el_i[1])*m.cos(beta)
        new_el_j = np.array([new_el_j_1, new_el_j_2])  
        
        
        population_s = np.concatenate([population_s, new_el_i])
        if abs(new_el_j[0]) <= 5 and abs(new_el_j[1]) <= 5:
            population_s = np.concatenate([population_s, new_el_j])
    population_s = np.reshape(population_s, (int(population_s.size/2),2))
    return population_s

#Step 6
def Generate_population_w(population):
    population_w = np.array([])
    while population_w.size < population.size:
        el_1 = rn.choice(population)
        el_2 = rn.choice(population)
        if fitness(el_1) < fitness(el_2):#если значение ф-ии в точке 1 меньше, чем в т. 2, то мы перезаписываем эту точку
            new_el_i = el_1
            el_j = el_2                  #в новую переменную, а вторую точку "пододвигаем" или же сжимаем к точке 1
        else:
            new_el_i = el_2
            el_j = el_1
        
        new_el_j_1 = [(el_j[0] + new_el_i[0])/k]
        new_el_j_2 = [(el_j[1] + new_el_i[1])/k]#коэф. сжатия
        new_el_j = np.concatenate([new_el_j_1, new_el_j_2])
        
        population_w = np.concatenate([population_w, new_el_i])
        if abs(new_el_j[0]) <= 5 and abs(new_el_j[1]) <= 5:
            population_w = np.concatenate([population_w, new_el_j])
    population_w = np.reshape(population_w, (int(population_w.size/2),2))
    return population_w
        
def condition(prev_population, cur_population):        
    prev = min(np.array(list(map(fitness, prev_population))))
    cur = min(np.array(list(map(fitness, cur_population))))# функция состояния популяции
    return abs(prev - cur) < eps
    
def main():
    population_t = Generate_population_t(number, dim)#генерация первой популяции
    prev_pop = population_t#нужно для того, чтобы заработал цикл while
    
    population_z = Generate_population_z(population_t)#генерация второй популяции
    population_s = Generate_population_s(population_t)#3
    population_w = Generate_population_w(population_t)#4
                                                # перезаписываем популяцию как конкатенацию предыдущих
    new_population_t = np.concatenate([population_t, population_z, population_s, population_w])
    f_new_population_t = np.array(list(map(fitness,new_population_t))).argsort()
    population_t = new_population_t[f_new_population_t[:number]]# берем из нее n "лучших" элементов  
    
    #в цикле все это повторяется
    while not condition(prev_pop, population_t):
        prev_pop = population_t
        population_z = Generate_population_z(population_t)
        population_s = Generate_population_s(population_t)
        population_w = Generate_population_w(population_t)

        new_population_t = np.concatenate([population_t, population_z, population_s, population_w])
        f_new_population_t = np.array(list(map(fitness,new_population_t))).argsort()
        population_t = new_population_t[f_new_population_t[:number]]
        f_population_t = np.array(list(map(fitness, population_t)))
        print('(x1, x2) = {}; F(x1, x2) = {}'.format(population_t[0], f_population_t[0]))
    print("--- %s seconds ---" % (time.time() - start_time))
        
    
        
        

#Test


func = main()
print('main', func,'\n')

```
